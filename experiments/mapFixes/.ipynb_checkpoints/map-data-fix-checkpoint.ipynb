```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.ops import unary_union
from shapely.affinity import scale
```

```python
upazilas = gpd.read_file('../../preprocessing/geodata/maps/dist/upa/upa-c005-s020-vw-pr.geojson').to_crs(epsg=32645)
unions = gpd.read_file('../../preprocessing/geodata/maps/dist/uni/uni-c005-s020-vw-pr.geojson').to_crs(epsg=32645)
mouzas = gpd.read_file('../../preprocessing/geodata/maps/dist/mou/mou-c005-s020-vw-pr.geojson').to_crs(epsg=32645)

print(upazilas.crs)
```

```python
print('========UPA========\n')
upazilas.info()
print('========UNI========\n')
unions.info()
print('========MOU========\n')
mouzas.info()
```

```python
upazilas['region_key'] = upazilas['div'] + '@' + upazilas['dis'] + '@' + upazilas['upa']
unions['region_key'] = unions['div'] + '@' + unions['dis'] + '@' + unions['upa'] + '@' + mouzas['uni']
mouzas['region_key'] = mouzas['div'] + '@' + mouzas['dis'] + '@' + mouzas['upa'] + '@' + mouzas['uni'] + '@' + mouzas['mou']
```

```python
upazilas = upazilas.sort_values(by='region_key')
unions = unions.sort_values(by='region_key')
mouzas = mouzas.sort_values(by='region_key')
```

```python
def number_region_keys(df):
    prev_region_key = ''
    region_key_index = 0
    region_key_indexes = []
    for i, row in df.iterrows():
        if row['region_key'] != prev_region_key:
            region_key_index += 1
            prev_region_key = row['region_key']
        region_key_indexes.append(region_key_index)
    return region_key_indexes

upazilas['region_key_index'] = number_region_keys(upazilas)
unions['region_key_index'] = number_region_keys(unions)
mouzas['region_key_index'] = number_region_keys(mouzas)
```

```python
# how many regions have duplicate keys
# rkivc == region key index value counts
morkivc = mouzas['region_key_index'].value_counts()
unrkivc = unions['region_key_index'].value_counts()
uprkivc = upazilas['region_key_index'].value_counts()

print('non unique mouza region keys ' + str(len(morkivc[morkivc > 1])))
print('non unique union region keys ' + str(len(unrkivc[unrkivc > 1])))
print('non unique upazila region keys ' + str(len(uprkivc[uprkivc > 1])))

print('\n')

mouzas_non_unique_count = morkivc[morkivc > 1].sum()
unions_non_unique_count = unrkivc[unrkivc > 1].sum()
upazilas_non_unique_count = uprkivc[uprkivc > 1].sum()

print(f"Number of rows with non-unique region_key_index in mouzas: {mouzas_non_unique_count}")
print(f"Number of rows with non-unique region_key_index in unions: {unions_non_unique_count}")
print(f"Number of rows with non-unique region_key_index in upazilas: {upazilas_non_unique_count}")

# after removing duplicates, we expect the number of polygons to be 
# (the starting number - the number of rows with non-unique region keys) + the num
```

```python
def find_containing_upazila(geometry, upazilas):
    for k, upazila in upazilas.iterrows():
        if upazila['geometry'].contains(geometry.centroid):
            return k
            
    upazilas['centroid'] = upazilas['geometry'].centroid
    upazilas['distance_to_geometry'] = upazilas['centroid'].apply(lambda x: geometry.centroid.distance(x))
    return upazilas.sort_values(by='distance_to_geometry').head(1).index[0]

ret_unions = unions.copy()
duplicate_keys = unrkivc[unrkivc > 1].index
df_arr = [ret_unions[ret_unions['region_key_index'] == key] for key in duplicate_keys]
merged_unions = []

for i, df in enumerate(df_arr):
    union_divs = set()
    union_diss = set()
    union_upas = set()

    containing_upa_ids = set()
    count = 0
    for j, geometry in df.iterrows():
        union_divs.add(geometry['div'])
        union_diss.add(geometry['dis'])
        union_upas.add(geometry['upa'])

        parent_upa_id = find_containing_upazila(geometry['geometry'], upazilas)
        containing_upa_ids.add(parent_upa_id)

        union_upa_key = '@'.join(geometry['region_key'].split('@')[:-1])
        if upazilas.loc[parent_upa_id]['region_key'] != union_upa_key:
            print(f'error processing {i + 1} / {len(df_arr)}')
            print('union upa key does not match upa region key')
            fig, ax = plt.subplots()
            gdf_containing_upazila = gpd \
                .GeoDataFrame(geometry=[upazilas.loc[parent_upa_id]['geometry']]) \
                .plot(ax=ax, linewidth=2, edgecolor='blue', facecolor='blue', alpha=0.5)

            gdf_labelled_upazila = gpd \
                .GeoDataFrame(geometry=upazilas.loc[upazilas['region_key'] == union_upa_key, 'geometry']) \
                .plot(ax=ax, linewidth=2, edgecolor='green', facecolor='green', alpha=0.5)
            
            gdf_union = gpd.GeoDataFrame(geometry=[geometry['geometry']])
            gdf_union.plot(ax=ax, linewidth=0.8, edgecolor='black', facecolor='white', alpha=0.5)
            plt.show()
            print('================================')
            
    merged_geometry = unary_union(df['geometry'])

    if len(union_divs) == 1 and len(union_diss) == 1 and len(union_upas) == 1 and len(containing_upa_ids) == 1:
        merged_unions.append({
            'region_key_index': df['region_key_index'].iloc[0],
            'region_key': df['region_key'].iloc[0],
            'uni': df['uni'].iloc[0],
            'upa': df['upa'].iloc[0],
            'dis': df['dis'].iloc[0],
            'div': df['div'].iloc[0],
            'geometry': merged_geometry,
            'area': df['area'].iloc[0],
        })

# for mu in merged_unions:
#     ret_unions = ret_unions[ret_unions['region_key_index'] != mu['region_key_index']]
#     new_row = pd.DataFrame({
#         'region_key_index': [mu['region_key_index']],
#         'region_key': [mu['region_key']],
#         'uni': [mu['uni']],
#         'upa': [mu['upa']],
#         'dis': [mu['dis']],
#         'div': [mu['div']],
#         'geometry': [mu['geometry']],
#         'area': [mu['area']],
#     })
#     ret_unions = pd.concat([ret_unions, new_row], ignore_index=True)

# print(ret_unions.info())

# how many unions have duplicate keys after fixes
unrkivc = ret_unions['region_key_index'].value_counts()
print('unions with duplicate region keys ' + str(len(unrkivc[unrkivc > 1])))
```

```python
def find_containing_union(geometry, unions, scale_factor=1.2):
    for k, union in unions.iterrows():
        if union['geometry'].contains(geometry.centroid):
            return k

    unions['centroid'] = unions['geometry'].centroid
    unions['distance_to_geometry'] = unions['centroid'].apply(lambda x: geometry.centroid.distance(x))
    return unions.sort_values(by='distance_to_geometry').head(1).index[0]

# Create a copy of mouzas dataframe
ret_mouzas = mouzas.copy()

# Find the duplicate keys in mouzas (based on region_key_index)
duplicate_keys = morkivc[morkivc > 1].index
df_arr = [ret_mouzas[ret_mouzas['region_key_index'] == key] for key in duplicate_keys]
merged_mouzas = []

for i, df in enumerate(df_arr):
    # print(f'processing {i + 1} of {len(df_arr)}')
    mouza_divs = set()
    mouza_diss = set()
    mouza_upas = set()

    containing_uni_ids = set()
    for j, geometry in df.iterrows():
        mouza_divs.add(geometry['div'])
        mouza_diss.add(geometry['dis'])
        mouza_upas.add(geometry['upa'])

        parent_uni_id = find_containing_union(geometry['geometry'], unions)
        containing_upa_ids.add(parent_uni_id)

        mouza_uni_key = '@'.join(geometry['region_key'].split('@')[:-1])
        # print(geometry['region_key'])
        # print(mouza_uni_key)
        # print(unions.loc[parent_uni_id]['region_key'])
        if unions.loc[parent_uni_id]['region_key'] == mouza_uni_key:
            # print(f'Error processing {i + 1} / {len(df_arr)}')
            print(f'processing {i + 1} of {len(df_arr)}')
            print('Mouza UPA key does match UPA region key')

            # fig, ax = plt.subplots()
            # gdf_containing_uni = gpd.GeoDataFrame(
            #     geometry=[unions.loc[parent_uni_id]['geometry']]
            # ).plot(ax=ax, linewidth=2, edgecolor='blue', facecolor='blue', alpha=0.5)

            # gdf_labelled_union = gpd.GeoDataFrame(
            #     geometry=unions.loc[unions['region_key'] == mouza_uni_key, 'geometry']
            # ).plot(ax=ax, linewidth=2, edgecolor='green', facecolor='green', alpha=0.5)

            # gdf_mouza = gpd.GeoDataFrame(geometry=[geometry['geometry']])
            # gdf_mouza.plot(ax=ax, linewidth=0.8, edgecolor='black', facecolor='black', alpha=0.2)
            # plt.show()
    #         print('--------------------------------')

    # print('================================')
    merged_geometry = unary_union(df['geometry'])

    if len(mouza_divs) == 1 and len(mouza_diss) == 1 and len(mouza_upas) == 1 and len(containing_upa_ids) == 1:
        merged_mouzas.append({
            'region_key_index': df['region_key_index'].iloc[0],
            'region_key': df['region_key'].iloc[0],
            'uni': df['uni'].iloc[0],
            'upa': df['upa'].iloc[0],
            'dis': df['dis'].iloc[0],
            'div': df['div'].iloc[0],
            'mou': df['mou'].iloc[0],
            'geometry': merged_geometry,
            'area': df['area'].iloc[0],
        })
print('done')
```
